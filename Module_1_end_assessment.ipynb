# Analysis Tasks:  


## Task 1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees. (2 marks)  


```python
# Step 1: Count employees in each team
team_counts = df['Team'].value_counts()
print(team_counts)
```

```python
# Step 2: Calculate percentage split
team_percentages = df['Team'].value_counts(normalize=True) * 100
print(team_percentages)
```

```python
# Combine into a summary DataFrame
team_summary = pd.DataFrame({
    'Employee Count': team_counts,
    'Percentage (%)': team_percentages.round(2)
})
print(team_summary)
```

## Task 1. Graphical representation

```python
team_counts.plot(kind='bar', color='b',width=.5, edgecolor='black', title='Employee Distribution by Team')
plt.ylabel('Number of Employees',color='g')
plt.xlabel('Team', color='g',)
plt.figure(figsize=(10, 6))
plt.show()
```

## Task 1: findings

* The New Orleans Pelicans have the highest employee count: 19 employees (4.15%).
* The Minnesota Timberwolves and Orlando Magic have the fewest: 14 employees each (3.06%).
* Most other teams have between 15 and 16 employees, indicating a balanced distribution overall.
* No major staffing imbalances are present, though small differences may reflect role specialization or regional needs.

## Task 3. Identify the predominant age group among employees. (2 marks) 

```python
print(min(df['Age']))
print(max(df['Age']))
```

```python
bins=[19, 24, 29, 34, 39, 44]
labels= ['19-23', '24-28','29-33','34-38','39-44']
age_groups = pd.cut(df['Age'],bins=bins,labels=labels,right=False )
predominant_age_group = age_groups.value_counts().idxmax()
print(predominant_age_group)
```

## Task 3 Graphical analysis

```python
age_groups.value_counts().plot(kind='bar')
plt.title('Age Group Distribution')
plt.ylabel('Number of Employees')
plt.show()
```

## Task 3: Findings

* Most employees fell into the 24-28 age group, followed by the 29-33 group. Very few were over 39.
* Insight: The workforce is relatively young, possibly due to the nature of the tech industry or a hiring preference for younger talent.

## Task 4. Discover which team and position have the highest salary expenditure. (2 marks)  

```python
salary_by_team = df.groupby('Team')['Salary'].sum()
salary_by_position = df.groupby('Position')['Salary'].sum()
highest_salary_team = salary_by_team.idxmax()
highest_salary_position = salary_by_position.idxmax()
print(F"Highest salary expenditure by team: {highest_salary_team}")
print(F"Highest salary expenditure by position: {highest_salary_position}")
```

## Task 4. Graphical Representation

```python
salary_by_team.plot(kind='bar')
plt.title('Salary Expenditure by Team')

plt.show()
salary_by_position.plot(kind='pie', autopct='%1.1f%%')
plt.title('Salary Expenditure by Position')
plt.show()
```

## Task 4. Findings

* The Cleveland Cavaliers team accounted for the highest total salary expenditure, and the PG position had the highest salary contribution among positions.
